```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from figure_grid import *
from transition_helpers import *
from modular_matrix import *
```

```python
A = standard_A(7,7,7)
Aeig = A - np.identity(49, dtype=np.int8)
Abasis = Aeig.solve(np.zeros(49), singular_mode="basis")
```

```python
with figure_grid(1, 7, rowwidth=20, rowheight=4) as grid:
    for i in range(Abasis.shape[0]):
        ax = grid.next_subplot()
        eig = Abasis[i, :].reshape((7,7))
        ax.imshow(eig, interpolation='none')
        ax.set_axis_off()
        for i in range(len(eig)):
            for j in range(len(eig[0])):
              plt.gca().text(i-0.175, j+0.175, int(eig[j][i]), fontsize=16, color='w')
```

```python
for i in range(len(Abasis)):
    print(np.array_equal((A*Abasis[i]).array, Abasis[i]))
```

```python
A = standard_A(7,7,7)
Aeig2 = A - 2*np.identity(49, dtype=np.int8)
Abasis2 = Aeig2.solve(np.zeros(49), singular_mode="basis")

with figure_grid(1, Abasis2.shape[0], rowwidth=20, rowheight=4) as grid:
    for i in range(Abasis2.shape[0]):
        ax = grid.next_subplot()
        eig = Abasis2[i, :].reshape((7,7))
        ax.imshow(eig, interpolation='none')
        ax.set_axis_off()
        for i in range(len(eig)):
            for j in range(len(eig[0])):
              plt.gca().text(i-0.175, j+0.175, int(eig[j][i]), fontsize=16, color='w')
```

```python
a1 = (2*Abasis2[0] % 7).reshape(7,7)
a2 = (A*Abasis2[0]).array.reshape(7,7)
np.array_equal(a1, a2)
```
